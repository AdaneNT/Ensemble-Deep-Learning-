# Comapre results of resampling approaches

```python
import pandas as pd
from pandas import read_csv
import matplotlib.pyplot as plt
import numpy as np
import numpy as np
import pandas as pd
```

```python
# read all the data 
df = pd.read_csv("WISDM_window_accuracy_2.csv",index_col=0)
print(df.shape)
df.head()
```

```python
df2=df
plt.figure(figsize=(10,5))
ax = df2.plot(kind='bar', figsize=(10,5))
plt.legend(loc='center left', bbox_to_anchor=(1, .7), fontsize='larger')
plt.grid()
ax.xaxis.set_tick_params(labelsize=15)
plt.xticks(rotation='horizontal')
ax.yaxis.set_tick_params(labelsize=10)
plt.xlabel('The data  after resampling approaches',size=16)
plt.ylabel('Imbalance Level  Measures',size=16)
#plt.title(' (a) ',size=15,color='black',weight="bold")
```
